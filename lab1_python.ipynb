# Tips for keys in Jupyter
## In command mode (blue colored cell)
 - Shift-Enter : execute and go to the next cell
 - Ctl-Enter : execute and stay
 - Alt-Enter : execute and create a new cell
 - a/b : create a new cell in above/below
 - x: delete cell
 - shift-m : merge two cells


## In edit mode (green colored cell)
- ctl-shift-"-" : split a cell

# Tips for keys in Colab
- Ctl-M + '/' : merge two cells
- Ctl-M + '-' : split a cell into two
- Ctl-M + 'M' : code -> text cell
- Ctl-M + 'Y' : text -> code cell

## coding example
- Monte Carlo simulation

```python
# method 1

import numpy as np

N = 100
x = np.random.rand(N)
y = np.random.rand(N)

cnt = 0
for i in range(N):
    if x[i]**2 + y[i]**2 < 1.0 :
        cnt = cnt + 1
    else:
        continue

pi = cnt / N * 4
print (pi)
```

```python

import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(4,4))
plt.scatter (x, y, s=2)

a = np.linspace(0, 1, 100)
b = (1 - a**2) ** 0.5
plt.plot(a, b, c='r')
```

```python
#   

counts = {}
with open("radishsurvey.txt") as file:
    for ilne in file:
        parts = line.strip()
        name, vote = parts.split(" - ")

        if vote not in counts:  # first appeared
            counts[vote] = 1
        else:                   # already exist, increment by 1
            counts[vote] = counts[vote] + 1
counts
```

```python
#  

counts = {}
with open("radishsurvey.txt") as file:
    for line in file:
        parts = line.strip()
        name, vote = parts.split(" - ")
        vote = vote.strip().capitalize().replace("  "," ")

        if vote not in counts:  # first appeared
            counts[vote] = 1
        else:                   # already exist, increment by 1
            counts[vote] = counts[vote] + 1
counts
```

```python
max_cnt = 0
for i in counts:
    if counts[i] > max_cnt:
        max_cnt = counts[i]
        max_radish = i
    else:
        continue
        
print("...Final result...")
print("max vote is ", max_radish, ": ", max_cnt)
```
